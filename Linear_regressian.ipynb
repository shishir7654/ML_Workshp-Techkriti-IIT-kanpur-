```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn import datasets , linear_model , metrics
from sklearn.model_selection import train_test_split
```

```python
# load the boston dataset
boston = datasets.load_boston(return_X_y = False)
X = boston.data
y = boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
print('cofficients:\n',reg.coef_)
```

```python
print('Variance, score:{}'.format(reg.score(X_test, y_test)))
```

```python
plt.style.use('fivethirtyeight')
```

```python
#plotting residual error in training data
plt.scatter(reg.predict(X_train),reg.predict(X_train)-y_train,color = "blue", s = 10,label = 'Train data')
plt.scatter(reg.predict(X_test),reg.predict(X_test)-y_test,color = "red", s = 10,label = 'Test data')
plt.hlines(y = 0,xmin = 0, xmax = 50, linewidth = 2)
## ploting legend
plt.legend(loc = 'upper right')
plt.title("Residual erroes")
plt.show()
```
